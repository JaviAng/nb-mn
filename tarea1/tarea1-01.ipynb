```python
#Cesar Andrés Galindo Villalobos - 161003313
#Juan Sebastian Correa Paez - 161003307

from time import time

bin = [0]

numero = float(input('Introduce el número a cambiar de base: '))

tiempo_inicial = time()

#Bit del signo
if numero < 0:
    bin[0] = 1
    numero = numero * -1.0
else:
    bin[0] = 0
        
#Separar parte interna y parte decimal 
parte_entera = int(numero)
parte_decimal = abs(numero) - abs(int(numero))

#Convertir número entero a binario

def binarizar(decimal):
    binario = ''
    while decimal // 2 != 0:
        binario = str(decimal % 2) + binario
        decimal = decimal // 2
    return str(decimal) + binario

#Correr la coma y guardar el número de veces que corrió

n=float(binarizar(parte_entera))

def digitos(n):
    if n > 9:
        n=n/10
        return 1+digitos(n)
    elif n <= 9:
        return 1
x = digitos(n)-1

i = 1
while i <= x:
    n = n/10
    i += 1
n=round(n,x) #Aproximar flotante

#Calcular exponente 8 bits

exp = x + 127

#Convertir exponente 8 bits a binario

def exponente(dec):
    binary = ''
    while dec // 2 != 0:
        binary = str(dec % 2) + binary
        dec = dec // 2
    return str(dec) + binary
ResulExp = exponente(exp)

#Calcular mantisa

mantisa = (abs(n) - int(abs(n)))
s = 1
while s <= x:
    mantisa = mantisa*10
    s += 1
    
def digitosMantisa(mantisa):
    if mantisa > 9:
        mantisa=mantisa/10
        return 1+digitos(mantisa)
    elif mantisa <= 9:
        return 1
    
z = digitosMantisa(mantisa)

mantisa=round(mantisa,x)
mantisa=int(mantisa)

while z < 23:
    mantisa = mantisa*10
    z += 1

mantisa=round(mantisa,2) #Aproximar flotante

#Mostrar binario de 32 bits

print('El número en formato IEEE 754 es: ',bin[0],ResulExp,int(mantisa))

tiempo_final = time() 
 
tiempo_ejecucion = tiempo_final - tiempo_inicial
 
print ('El tiempo de ejecucion fue:',tiempo_ejecucion) #En segundos
```
