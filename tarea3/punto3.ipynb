```python
def f(x):
    return ( 4*x**4  + 9*x**3 - 5*x**2 + 9*x**1 - 9)

tolerancia = 0.001

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

puntos = int(round(1/tolerancia))
x = np.linspace(-5, 5, puntos)
y = f(x)

plt.plot(x, y, lw=1)
plt.grid(True, linestyle='-.')


a = 0
b =4
c = (a + b) / 2.0

while (b - a) / 2.0 > tolerancia:
    print ("{:f} {:f} {:f}".format(a, c, b))
    if f(c) == 0:
        break
    elif f(a) * f(c) < 0:
        b = c
    else:
        a = c
    c = (a + b) / 2.0


print("La raiz aprox. es: ")
print(c)

#el punto 3 tiene dos raices, en -3 y o.75
```
