```python
# Ecuacion 1 
import numpy 
def f(x):
    return (3*x + numpy.sin(x)- numpy.exp(x))

tolerancia = 0.001
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

puntos = int(round(1/tolerancia))
x = np.linspace(-1,2.5, puntos)
y = f(x)


plt.plot(x, y, lw=2)
plt.grid(True, linestyle='-.')

a = 0
b = 1.0
c = (a + b) / 2.0

while (b - a) / 2.0 > tolerancia:
   # print "{:f} {:f} {:f}".format(a, c, b)
    if f(c) == 0:
        break
    elif f(a) * f(c) < 0:
        b = c
    else:
        a = c
    c = (a + b) / 2.0
    
    
print("-------------------------")
print("La raiz numero 1. es: ")
print(c)

#-----------------------------------------------------------------------------------------------------------
print("")
a = 1.0
b = 3.0
c = (a + b) / 2.0

while (b - a) / 2.0 > tolerancia:
  #  print "{:f} {:f} {:f}".format(a, c, b)
    if f(c) == 0:
        break
    elif f(a) * f(c) < 0:
        b = c
    else:
        a = c
    c = (a + b) / 2.0
    
    
print("-------------------------")
print("La raiz numero 2. es: ")
print(c)

print("               Grafica de la funcion  ")
```
