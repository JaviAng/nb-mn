```python
"F 2"

import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 5/np.sqrt(x-(1/x))

x = np.linspace(-8 , 8 , 101)
plt.plot(x,f(x))
plt.grid()
plt.show()

a = -9;
b = 8;
raiz = 0;
tolerancia = 8
for i in range (40):
    if tolerancia > 0.0001 :
        c = (a+b)/2;
        fa = f(a);
        fc = f(c);
        if (fc == 0):
            raiz = c
            break
        elif (fa * fc < 0 ):
            b = c;
        else :
            a = c;
            raiz = c;
            tolerancia = abs(fc);

print("La raiz de la funcion es:",raiz)
```
