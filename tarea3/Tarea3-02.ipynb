```python
## punto 1##

def f(x):
    return 3*x + np.sin(x) - np.exp(x)

tolerancia = 0.001

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

puntos = int(round(1/tolerancia))
x = np.linspace(-2, 2, puntos)
y = f(x)

plt.plot(x, y, lw=2)
plt.grid(True, linestyle='-.')

a = -2
b = 2
c = (a + b) / 2.0


while (b - a) / 2.0 > tolerancia:
    print ("{:f} {:f} {:f}".format(a, c, b))
    if f(c) == 0:
        break
    elif f(a) * f(c) < 0:
        b = c
    else:
        a = c
    c = (a + b) / 2.0

print("La raiz aprox. es: ")
print(c)
```
