```python
from numpy import tan
from numpy import sin
from numpy import cos
from numpy import exp
from numpy import sqrt
from numpy import log
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from __future__ import division

def f1(x):
    return 3*x-sin(x)-exp(x)
def f2(x):
    return 5/sqrt(x-(1/x))
def f3(x):
    return (4*(x**4))+(9*(x**3))-(5*(x**2))+(9*x)-9
def f4(x):
    return exp(x)+sqrt(x)-2
def f5(x):
    return (2*(x**3))-(x**2)-(8*x)+4
def f6(x):
    return 1/((x-2)*((x**2)+(3*x)+1))
def f7(x):
    return exp((x**3)+2*(x**2)+x+1)
def f8(x):
    return (log(x)+(1/((x**2)-5)*(x-2)))
def f9(x):
    return 3*(x**2)+log(5*x)/(3*(x**3))-(1/(exp(2)*x+3))
def f10(x):
    return 4*cos(log(x))
def f11(x):
    return (x+2)/(5*(x**2)+12*x+4)
def f12(x):
    return 4*(x**2)+4*x+17
def f13(x):
    return (cos(x)+x)/(sin(x+exp(sqrt(tan(x)))))

tol=0.0001
puntos = int(round(1/tol))

x = np.linspace(-5, 6, puntos)
y = f1(x)
y1 = f2(x)
y2 = f3(x)
y3 = f4(x)
y4 = f5(x)
y5 = f6(x)
y6 = f7(x)
y7 = f8(x)
y8 = f9(x)
y9 = f10(x)
y10 = f11(x)
y11 = f12(x)
y12 = f13(x)

plt.subplot(2, 2, 1)
plt.plot(x, y, lw=1)
plt.grid(True, linestyle='-.')
plt.title('f1')

plt.subplot(2, 2, 2)
plt.plot(x, y1, lw=1)
plt.grid(True, linestyle='-.')
plt.title('f2')

plt.subplot(2, 2, 3)
plt.plot(x, y2, lw=1)
plt.grid(True, linestyle='-.')
plt.title('     f3')

plt.subplot(2, 2, 4)
plt.plot(x, y3, lw=1)
plt.grid(True, linestyle='-.')
plt.title('f4')

plt.show()
```
