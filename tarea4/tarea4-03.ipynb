```python
print(" TAREA 4. PRONOSTICO DE UN COLAPSO")
print(" BRAYAN J. SALDARRIAGA 161003640")
print(" FELIPE CORAL OTALORA 161003611")
import math
archivo= open('datos.txt','r')
entrada=archivo.read()
datos= str(entrada).split(",")
#subindice 0 es para la nave, subindice i es para el planeta
mi=float(datos[0])
m0=float(datos[1])
r=float(datos[2])
xi=float(datos[3])
yi=float(datos[4])
x0=float(datos[5])
y0=float(datos[6])
vx=float(datos[7])
vy=float(datos[8])
colapso=0
radio1=(((x0-xi)**2)+((y0-yi)**2))**(1/2)
radio=(((x0-xi)**2)+((y0-yi)**2))**(1/2)
constante=6.674*(10**-11)
gravedad=(constante*((mi*m0)/(radio**(2))))
tiempo=0
print(radio)
print("distancia entre el centro de la masa 1 al de la masa 2: ",radio1)
print("fuerza de atraccion gravitaroria de la masa1 sobre la masa 2: ",gravedad)
while(radio>r and radio<2*radio1):
    xt=x0-xi
    yt=y0-yi
    print("posicion inicial ",yt,xt)
    ang= math.atan(xt/yt)
    print("angulo ", ang)
    fx=gravedad*math.sin(ang)
    fy=gravedad*math.cos(ang)
    vfx=(fx/m0)+vx
    vfy=(fy/m0)+vy
    print("velocidad final ",vfx," ",vfy)
    x0=x0+vfx
    y0=y0+vfy
    print("posicion final ",x0," ",y0)
    radio=(((x0-xi)**2)+((y0-yi)**2))**(1/2)
    print("radio inicial ",radio1," radio final ",radio)
    tiempo=tiempo+1
    print ("tiempo ", tiempo)
if (radio<=r):
    print("la nave colapso despues de ",tiempo," unidades de tiempo")
elif(radio>=(2*radio1)):
    print("la nave se salvó")
```
