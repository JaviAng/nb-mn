```python
import math
import matplotlib.pyplot as plt
import numpy as np

print("urley camilo guerrero rodriguez 161003449 ricardo ivan diaz moreno 161003415")
print("tarea#4")

archivo= open('text.txt','r')
In=archivo.read()
data= str(In).split(",")
m1 = float(data[0])
m2 = float(data[1])
r = float(data[2])
x1 = float(data[3])
y1 = float(data[4])
x2 = float(data[5])
y2 = float(data[6])
vx = float(data[7])
vy = float(data[8])

G=6.674*(10**-11)

#posicion del cohete(púnto)
#vector velocidad del cohete

vi=vy/vx

#ecuacion forma punto-pendiente del planeta y la nave, distancia casi de separacion 
#y = yo + b/a ( x - xo ) 


a=vi
b=y1-(vi*x1)
B=(b)/((a**2)+1)**(1/2)

#magnitud velocidad de la nave
vo=((vx**2)+(vy**2))**(1/2)

#distancia de la nave con el radio del planeta
E=(1+((B**2)*(vo**4))/((m2**2)*(G)))**(1/2)
D=((B**2)*(vo**2))/((m2)*(G))
rp=(D)/(1+E) 
RP=rp/100

#Fuerza gravitacional entre los dos objetos
#C=(((G)*(m1*m2)/(r**2))

#ecuacion de la recta para la distancia entre los radio del planeta y la nave
#a = vy/vx
#b = y1 - ( a * x1 ) #distancia entre el recorrido de la nave y el planeta


#d = ((a*x1)-y1+b)/(((a**2)+1)**0.5) #distancia del radio del planeta con respecto a la nave
#vo=((vx**2)+(vy**2))**(1/2) #magnitud de la velicidad inicial de la nave

#ecuacions colision de la nave
#E=(1+((b**2)*(vo**4))/((m2**2)*(G)))**(1/2)
#D=((b**2)*(vo**2))/((m2)*(G))
#rp=(D)/(1+E) #distancia de la nave con el radio del planeta

#radio entre la nave y el planeta
#r=(((x1-x2)/(2))+((y1-y2)**2)**2)

print("velocidad cohete eje x:",vx)
print("velocidad cohete eje y:",vy)
print("radio del planeta:",r)
print("ubicacion del cohete eje x:",x1)
print("ubicacion del cohete eje y:",y1)
print("ubicacion del planeta eje x:",x2)
print("ubicacion del planeta eje y:",y2)
print("masa del planeta:",m2)
print("ubicacion del cohete:",m1)
  
if  rp <= r: #radio de la nave en el planeta con respecto al radio del planeta
    print("la nave se estrello:0")
else:
    print("la nave no se estrello:1")
    
print("radio de la nave contra el planeta:",rp)
print("radio del planeta:",r)
print("distancia entre en centro del planeta y la nave:",B)
print("Codigo sujeto a cambios")

X = np.linspace(-r,r,3000)
Y = np.sqrt(-X**2+r**2)
#plt.plot(X, Y,'b')
#plt.plot(X,-Y,'b')
#plt.gca().set_aspect('equal')
#plt.show()

X1 = np.linspace(-RP,RP,2500)
Y1 = np.sqrt(-X1**2+RP**2)

#def f(rp):   #prueba grafica nave#1
#   return np.exp(rp**0.5)
#rp = np.linspace(-50,50 , num=5000)

fig, ax = plt.subplots()
ax.plot(X, Y, label="radio",color="red")
ax.plot(X, -Y,color="red")
ax.plot(X1,Y1, label="nave")
ax.legend();
plt.show()
```
