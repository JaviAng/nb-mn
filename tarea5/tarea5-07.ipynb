## La posicion del del nuevo centro de masa luego de la explosion 
\begin{equation*}
R =
\frac{(qm_1)r_1 + m_2r_2}{(qm_1 + m_2)}
\end{equation*}

## Energia cinetica de la explosion binaria
\begin{equation*}
E =
\frac{1}{2}mv^2
\frac{(q(m_1-m_2-2qm_1)}{(qm_1 + m_2)}
\end{equation*}

## La energía total del movimiento relativo del sistema binario posterior a la explosión
\begin{equation*}
E =
\frac{1}{2}
\frac{qm_1r_1m_2}{(qm_1 + m_2)} v^2 -
\frac{Gqm_1m_2}{a_0}
\end{equation*}

## Energía cinetica para el movimiento relativo posterior a la explosión binaria
\begin{equation*}
E =
\frac{1}{2}(qm_1)v_1^2 + 
\frac{1}{2}(m_2v_2^2)- 
\frac{1}{2}(qm_1 + m_2)R^2- G(qm_1)
\frac{m_2}{a_0}
\end{equation*}

## Velocidad relativa del binario previo a la explosión
\begin{equation*}
v^2 = 
\frac{G(m_1 + m_2}{a_0}
\end{equation*}

## La energía total del binario posterior a la explosión se puede reducir a
\begin{equation*}
E =
\frac{1}{2}
\frac{mv^2[q(m_1 - m_2 - 2qm_1)]}{qm_1 + m_2}
\end{equation*}

## Velocidad despues de la explosion
La velocidad de el centro de masa despues de la explosion

\begin{equation*}
V_{cm} =
\frac{(1-q)m_2v_2}{qm_1 + m_2} 
\end{equation*}

\begin{equation*}
V_{cm} =
\frac{m(1-q)}{(qm_1 + m_2)}
\frac{G(m_1+m_2)}{a_0}^{1/2}
\end{equation*}

La velocidad relativa de dos componentes de un binario sin consolidar

\begin{equation*}
V=
\frac{G(m_1 - m_2 - 2qm_1}{a_0}^{1/2} =
\frac{e-1}{e+1}^{1/2}
\end{equation*}

Velocidad minima relativa de un binario consolidado

\begin{equation*}
V{min}=[
\frac{(1-e)}{(1+e)^{1/2}} ]v
\end{equation*}

\end{equation*}

\begin{equation*}
\frac{d}{dt} 
\frac{dy_1}{dt} =

# Graficos de un sistema binario de orbitas cilindricas y elipticas

A continuacion mostramos un codigo simple, para visualizar el comportamiento de las orbitas en los sistemas binarios.

```python
import numpy as np
from matplotlib import patches
import matplotlib.pyplot as plt
```

```python
#Masa de las estrellas
m_1=20
m_2=5

#Masa de la estrella 1 al perder el 20% de su masa
m_11=m_1*0.8

#El radio entre las estrellas es normalizado y multiplicado por 10
r=10

r_1=m_2*r/(m_1+m_2)
r_2=m_1*r/(m_1+m_2)

r_11=m_2*r/(m_11+m_2)


#Graficas de orbitas circulares
fig = plt.figure()

ax = fig.add_subplot(211, aspect='equal')
ax.axis([-12, 12, -12, 12])
ax.text(-11,-10,"Normal")

c1 = patches.Circle((0,0), radius=r_1,fill=False)
c2 = patches.Circle((0,0), radius=r_2,fill=False)
c11 = patches.Circle((-r_1,0), radius=m_1/10)
c22 = patches.Circle((r_2,0), radius=m_2/10)

ax.add_patch(c1)
ax.add_patch(c2)
ax.add_patch(c11)
ax.add_patch(c22)

ax = fig.add_subplot(212, aspect='equal')
ax.axis([-12, 12, -12, 12])
ax.text(-11,-10,"Reduccion masa")

c3 = patches.Circle((0,0), radius=r_11,fill=False)
c4 = patches.Circle((0,0), radius=r_2,fill=False)
c33 = patches.Circle((-r_11,0), radius=m_11/10)
c44 = patches.Circle((r_2,0), radius=m_2/10)

ax.add_patch(c3)
ax.add_patch(c4)
ax.add_patch(c33)
ax.add_patch(c44)

fig.savefig('circle_compare')
```

```python
fig = plt.figure()

ax = fig.add_subplot(211, aspect='equal')
ax.axis([-12, 12, -12, 12])
ax.text(-11,-10,"Normal")

e1 = patches.Ellipse((-1, 0), 5, 3, fill=False)
e2 = patches.Ellipse((1, 0), 4, 2, fill=False)
c11 = patches.Circle((-r_1,0), radius=m_1/10)
c22 = patches.Circle((r_2,0), radius=m_2/10)

ax.add_patch(e1)
ax.add_patch(e2)
ax.add_patch(c11)
ax.add_patch(c22)

ax = fig.add_subplot(212, aspect='equal')
ax.axis([-12, 12, -12, 12])
ax.text(-11,-10,"Reduccion masa")

e1 = patches.Ellipse((-1, 0), 5, 3, fill=False)
e2 = patches.Ellipse((1, 0), 7, 2.5, fill=False)
c33 = patches.Circle((-r_11,0), radius=m_11/10)
c44 = patches.Circle((r_2,0), radius=m_2/10)

ax.add_patch(e1)
ax.add_patch(e2)
ax.add_patch(c33)
ax.add_patch(c44)

fig.savefig('ecllipse_compare')
```
